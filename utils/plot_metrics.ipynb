```python
import pandas as pd
import matplotlib.pyplot as plt

# Задаем переменные
path_to_log = '/home/super/mironov/mia/logs/metrics.csv'
path_to_lr_log = '/home/super/mironov/mia/logs/lr.csv'

# Загружаем данные
log_df = pd.read_csv(path_to_log)
lr_df = pd.read_csv(path_to_lr_log)
```

```python
metric = 'bal_acc' #'loss'  # Пример: 'loss', 'recall', 'precision', 'bal_acc', 'acc'
set_type = 'valid'  # Пример: 'train', 'valid', 'test'

# Фильтруем по типу сета
filtered_df = log_df[log_df['set'] == set_type]

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['step'], filtered_df[metric], marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel(metric.capitalize())
plt.title(f'{metric.capitalize()} over Steps ({set_type} set)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lr_df['epoch'], lr_df['lr'], marker='o', linestyle='-')
plt.xlabel('Epoch')
plt.ylabel('Learning rate')
plt.title('Learning rate over Epochs')
plt.grid(True)
plt.show()
```
